## Задание 3. Выделить стратегию, которая кажется наименее приемлемой для себя, и написать список из 7-8 ее преимуществ.  






### Несмотря на результаты теста, я считаю для себя наиболее приемлемым в рабочих взаимоотношениях использовать стратегию компромисса по следующим причинам:


1. Необходимость сохранения хороших отношений между оппонентами после удачного разрешения спорной ситуации.  



2. Получение дополнительного опыта в области проведения переговоров, приобретение умений и навыков по успешному разрешению конфликтов.



3. Минимизация возможных рисков, и, как следствие, невозможность дальнейшего обострения ситуации.



4. Возможность окончательно разрешить давно назревающий конфликт и больше никогда не возвращаться к нему.



5. При компромиссе каждый уходит с чем-то, чем можно похвалиться за свои усилия, т.к. никому не нравится уходить от конфликта с чувством, что они ничего не выиграли. 


6. Повышение эмоциональной стабильности, т.к. пойдя на компромисс, можно сохранить свою эмоциональную устойчивость и избежать негативного влияния на свое психическое состояние. 


7. Многие люди уважают тех, кто готов идти на компромисс. Во многих случаях такая готовность рассматривается как признак зрелости и желания быть командным игроком. 


8. Если работаешь в команде необходимо воспринимать доводы и мнения другой стороны как свои собственные, относиться к ним с уважением.
